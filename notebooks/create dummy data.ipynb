```python
import numpy as np
import pandas as pd
import datetime
```

```python
start = datetime.datetime(year=2014, month=7, day=23, hour=4)
end = datetime.datetime(year=2014, month=7, day=24, hour=4)
# dt = datetime.timedelta(minutes=5)
# times = [begin + i*dt for i in range()]
```

```python
times = pd.date_range(start=start, end=end, freq="5s", closed='left')
```

```python
awake_hour = np.hstack((np.zeros(9), np.ones(22-9), np.zeros(3)))
```

```python
samples_per_hour = 60 * 60/5
```

```python
awake = np.hstack((np.zeros(8 * samples_per_hour),
                    np.ones((22-8) * samples_per_hour),
                    np.zeros(2 * samples_per_hour)))
```

```python
df = pd.DataFrame(data=np.array((awake,)).T, index=times, columns=['awake'])
# df.resample('5s')
# df['datetime'] = times.to_datetime()
```
